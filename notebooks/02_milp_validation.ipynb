# MILP Backtest on Real Bangalore Data

Loads `data/raw/icmc.json`, replays the past 6 months under both FIFO baseline
and our MILP optimizer, prints the comparison table that goes on the pitch deck.

```python
import json, sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent / 'apps' / 'api'))
from nagarik.milp.cvrptw import solve_cvrptw, naive_fifo_baseline
from nagarik.milp.backtest import synth

payload = synth(n_issues=150, n_crews=10)
fifo = naive_fifo_baseline(payload)
milp = solve_cvrptw(payload)

print('FIFO :', fifo['metrics'])
print('MILP :', milp['metrics'])
print('km reduction:', round(100 * (1 - milp['metrics']['total_km'] / fifo['metrics']['total_km']), 1), '%')
```
